```python
import os

# This looks at the folder where your notebook is saved
start_looking_here = os.getcwd() 

print(f"🔎 Searching for the model starting from: {start_looking_here}\n")

found_anything = False
for root, dirs, files in os.walk(start_looking_here):
    for file in files:
        if file.endswith(".h5") or file.endswith(".keras") or file.endswith(".pkl"):
            full_path = os.path.join(root, file)
            print(f"FOUND: {file}")
            print(f"EXACT PATH: r'{full_path}'\n")
            found_anything = True

if not found_anything:
    print("❌ Still nothing. The files might be outside this folder or not created yet.")
```

```python
import tensorflow as tf
import joblib
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 1. THE EXACT PATHS (Fixed to point to the actual files)
MODEL_FILE = r'C:\Users\vikas\OneDrive\Documents\Phishing-Email-Detection-System\backend\backend\models\cnn_phishing_model.h5'
TOKEN_FILE = r'C:\Users\vikas\OneDrive\Documents\Phishing-Email-Detection-System\backend\backend\models\tokenizer.pkl'

# 2. LOAD ASSETS
try:
    # This should now work because we are pointing to the .h5 file, not the folder
    model = tf.keras.models.load_model(MODEL_FILE)
    tokenizer = joblib.load(TOKEN_FILE)
    print("✅ SUCCESS: 23MB Brain is loaded and thinking!")
except Exception as e:
    print(f"❌ Still an error: {e}")

# 3. THE TESTER
def test_email(text):
    # Standard cleaning
    clean = text.lower()
    seq = tokenizer.texts_to_sequences([clean])
    padded = pad_sequences(seq, maxlen=300, padding='post', truncating='post')
    
    # AI Prediction
    prob = float(model.predict(padded, verbose=0)[0][0])
    
    print(f"\nScanning: {text[:60]}...")
    print(f"Phishing Probability: {prob:.4f}")
    print("Verdict:", "🚨 PHISHING" if prob > 0.5 else "✅ SAFE")

# 4. RUN A SCAN
test_email("URGENT: Your account password expires in 2 hours. Click http://secure-login.net to update.")
```
